```
# Define the columns and the aggregations to apply
columns_to_agg = ['Age', 'Academic Pressure', 'Study Satisfaction', 'Study Hours', 'Financial Stress']
aggregations = ['sum', 'min', 'max']

# Create a dictionary with column names as keys and the list of aggregations as values
agg_dict = {col: aggregations for col in columns_to_agg}

# Apply the aggregation
display(df.agg(agg_dict))
```

```
pivot_table = df.pivot_table(index='Gender', values=['Age', 'Academic Pressure', 'Study Satisfaction', 'Study Hours', 'Financial Stress'], aggfunc='sum')
display(pivot_table)
```

```
melted_data = df.melt(id_vars=['Gender'], value_vars=['Age', 'Academic Pressure', 'Study Satisfaction', 'Study Hours', 'Financial Stress'], var_name='Metric', value_name='Value')
display(melted_data)
```
